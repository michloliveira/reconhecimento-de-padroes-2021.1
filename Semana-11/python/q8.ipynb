```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread("./imagens/Fig0335(a)(ckt_board_saltpep_prob_pt05).png", cv2.IMREAD_GRAYSCALE)
img = img.astype(np.uint8)

img.shape


plt.imshow(img ,'gray')
plt.axis("off")

plt.show()
```

8. (10 pontos) Compare os efeitos de um  filtro de mdia 3  3 e de um  filtro de mediana 3  3 na imagem  Fig0335(a)(ckt_board_saltpep_prob_pt05) . 

```python
media3x3 = cv2.blur(img, (15,15)).astype(np.uint8)
mediana3x3 = cv2.medianBlur(img, 3).astype(np.uint8)
```

```python
plt.subplot(1,2,1)
plt.imshow(media3x3 ,'gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(mediana3x3 ,'gray')
plt.axis("off")

plt.show()
```
