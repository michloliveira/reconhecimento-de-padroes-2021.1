```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread("./imagens/Fig0334(a)(hubble-original).png", cv2.IMREAD_GRAYSCALE)
img = img.astype(np.uint8)

img.shape
```

```python
plt.imshow(img ,'gray')
plt.axis("off")

plt.show()
```

(a) Aplique um  filtro de m√©dia 15x15. 

```python
img15x15 = cv2.blur(img, (15,15)).astype(np.uint8)


plt.subplot(1,2,1)
plt.imshow(img ,'gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(img15x15 ,'gray')
plt.axis("off")

plt.show()
```

(b) Sobre as sa√≠da da etapa anterior, limiarize a imagem utilizando limiar 64

```python
img15x15.shape
```

```python
imgFinal = img.astype(np.uint8)

for i in range(img15x15.shape[0]):
    for j in range(img15x15.shape[1]):
        if img15x15[i][j] > 64:
            imgFinal[i][j] = 255
        else:
            imgFinal[i][j] = 0
```

```python
plt.subplot(1,2,1)
plt.imshow(img ,'gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(imgFinal ,'gray')
plt.axis("off")

plt.show()
```

(c) Multiplique a sa√≠da da quest√£o anterior pela imagem original. 

```python
imgFinalMult = img.astype(np.uint8)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        imgFinalMult[i][j] = int(imgFinal[i][j]) * int(img[i][j])
```

```python
plt.subplot(1,2,1)
plt.imshow(img ,'gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(imgFinalMult ,'gray')
plt.axis("off")

plt.show()
```
