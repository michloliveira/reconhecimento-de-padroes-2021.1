```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread("./imagens/Fig1039(a)(polymersomes).png", cv2.IMREAD_GRAYSCALE)
img = img.astype(np.uint8)

img.shape
```

```python
limiar,otsuTotal = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.imshow(otsuTotal ,'gray')
plt.axis("off")

plt.show()
```

```python
topLeft = img[0:234,0:216].astype(np.uint8)
topMid = img[0:234, 216:432].astype(np.uint8)
topRight = img[0:234, 432:648].astype(np.uint8)

midLeft = img[234:468,0:216].astype(np.uint8)
midMid = img[234:468, 216:432].astype(np.uint8)
midRight = img[234:468, 432:648].astype(np.uint8)

bottonLeft = img[468:702,0:216].astype(np.uint8)
bottonMid = img[468:702, 216:432].astype(np.uint8)
bottonRight = img[468:702, 432:648].astype(np.uint8)

```

```python
limiar,topRightOut = cv2.threshold(topRight,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
limiar,topMidOut = cv2.threshold(topMid,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
limiar,topLeftOut = cv2.threshold(topLeft,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

limiar,midRightOut = cv2.threshold(midRight,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
limiar,midMidOut = cv2.threshold(midMid,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
limiar,midLeftOut = cv2.threshold(midLeft,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

limiar,bottonRightOut = cv2.threshold(bottonRight,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
limiar,bottonMidOut = cv2.threshold(bottonMid,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
limiar,bottonLeftOut = cv2.threshold(bottonLeft,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
plt.subplot(3,3,1)
plt.imshow(topLeftOut ,'gray')
plt.axis("off")
plt.subplot(3,3,2)
plt.imshow(topMidOut ,'gray')
plt.axis("off")
plt.subplot(3,3,3)
plt.imshow(topRightOut ,'gray')
plt.axis("off")
plt.subplot(3,3,4)
plt.imshow(midLeftOut ,'gray')
plt.axis("off")
plt.subplot(3,3,5)
plt.imshow(midMidOut ,'gray')
plt.axis("off")
plt.subplot(3,3,6)
plt.imshow(midRightOut ,'gray')
plt.axis("off")
plt.subplot(3,3,7)
plt.imshow(bottonLeftOut ,'gray')
plt.axis("off")
plt.subplot(3,3,8)
plt.imshow(bottonMidOut ,'gray')
plt.axis("off")
plt.subplot(3,3,9)
plt.imshow(bottonRightOut ,'gray')
plt.axis("off")
plt.show()
```
