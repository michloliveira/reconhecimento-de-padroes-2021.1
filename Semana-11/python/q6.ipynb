```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread("./imagens/Fig0333(a)(test_pattern_blurring_orig).png", cv2.IMREAD_GRAYSCALE)
img = img.astype(np.uint8)

img.shape
```

```python
plt.imshow(img ,'gray')
plt.axis("off")

plt.show()
```

```python
img3x3 = cv2.blur(img, (3,3))

plt.subplot(1,2,1)
plt.imshow(img ,'gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(img3x3 ,'gray')
plt.axis("off")

plt.show()
```
