```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python

img = cv2.imread("./imagens/Fig1039(a)(polymersomes).png", cv2.IMREAD_GRAYSCALE)
img = img.astype(np.uint8)
T = 173
newT = 0
delta = 5

while True:
    media1 = []
    media2 = []
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i][j] > T:
                media1.append(img[i][j])
            else:
                media2.append(img[i][j])
    

    newT = ((np.mean(media1)) + np.mean(media2))/2

    if (T - newT) < delta:
        break
    T = newT

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i][j] < T:
            img[i][j] = 0
        else:
            img[i][j] = 255

plt.imshow(img, 'gray')
plt.axis("off")
plt.show()

    

```
