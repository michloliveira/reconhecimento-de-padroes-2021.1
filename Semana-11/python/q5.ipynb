```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread("./imagens/Fig1036(c)(gaussian_noise_mean_0_std_50_added).png", cv2.IMREAD_GRAYSCALE)
img = img.astype(np.uint8)

img.shape
```

```python
plt.imshow(img ,'gray')
plt.axis("off")

plt.show()
```

```python
limiar,imgSemProcess = cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.imshow(imgSemProcess ,'gray')
plt.axis("off")

plt.show()
```

Com processamento

```python
img2 = cv2.blur(img, (5,5))

plt.imshow(img2 ,'gray')
plt.axis("off")

plt.show()
```

```python
limiar,imgComProcess = cv2.threshold(img2,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python

plt.imshow(imgComProcess ,'gray')
plt.axis("off")

plt.show()
```
