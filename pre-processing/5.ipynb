```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
```

```python
def mediaAmostral(accuracyArray):
    return np.mean(accuracyArray)
```

```python
def desvioPadraoAmostral(accuracyArray):
    return np.std(accuracyArray, ddof = 1)
```

```python
def intervaloConfiancaTaxaAcerto(accuracy):
    #media - 1,96 x desvioPadrao ; media  + 1,96 x desvioPadrao
    media = mediaAmostral(accuracy)
    desvio = desvioPadraoAmostral(accuracy)

    print("[{0:.2f};{1:.2f}]".format((media - 1.96 * desvio), (media + 1.96 * desvio)))
```

```python
df = pd.read_csv("wine.data", header=None)
df
```

```python
accuracy0 = []
accuracy1 = []
x = df[df.columns[1::]]
y = df[df.columns[0]]
```

```python
for i in range(100):
    treinoX, testeX, treinoY, testeY = train_test_split(x,y,test_size=0.50)
    knn = KNeighborsClassifier(n_neighbors=1, weights= "distance",metric="euclidean")
    knn.fit(treinoX,treinoY)
    y_pred = knn.predict(testeX)
    y_true = testeY
    #print(accuracy_score(y_true,y_pred))
    accuracy0.append(accuracy_score(y_true,y_pred))

intervaloConfiancaTaxaAcerto(accuracy0)
```
