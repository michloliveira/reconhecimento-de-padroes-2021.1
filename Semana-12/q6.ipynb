```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("imagens/Fig1016(a)(building_original).png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.axis("off")
plt.show()
```

```python
x = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
```

```python
y = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
```

(a) utilizando o gradiente diretamente na imagem;

```python
final = cv2.addWeighted(x, 2, y, 2, 0)
plt.imshow(final, 'gray')
plt.axis("off")
plt.show()
```

(b) utilizando o gradiente na imagem suavizada com um filtro de média 5x5;

```python
imgSmoothd = img2 = cv2.blur(img, (5,5))
plt.imshow(imgSmoothd, cmap='gray')
plt.axis("off")
plt.show()
```

```python
xSmoothd = cv2.Sobel(imgSmoothd, cv2.CV_64F, 1, 0, ksize=3)
ySmoothd = cv2.Sobel(imgSmoothd, cv2.CV_64F, 0, 1, ksize=3)
finalSmoothd = cv2.addWeighted(xSmoothd, 2, ySmoothd, 2, 0)

plt.imshow(finalSmoothd, 'gray')
plt.axis("off")
plt.show()
```

(c) utilizando um limiar 80 no resultado da letra (a);

```python
imgLimiarA = final.astype(np.uint8)

for i in range(final.shape[0]):
    for j in range(final.shape[1]):
        if final[i][j] > 80:
            imgLimiarA[i][j] = 255
        else:
            imgLimiarA[i][j] = 0

plt.imshow(imgLimiarA, cmap='gray')
plt.axis("off")
plt.show()
```
