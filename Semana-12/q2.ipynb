```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("imagens/Fig0340(a)(dipxe_text).png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, 'gray')
plt.axis("off")
plt.show()
```

```python
k = 2
resultant_img = img.copy()
for i in range(1,img.shape[0]-1):
        for j in range(1,img.shape[1]-1):
            blur_factor = cv2.GaussianBlur(img, (5, 5), 3)
            mask = int(img[i,j]) - int(blur_factor[i,j])
            resultant_img[i,j] = int(img[i,j]) + k * int(mask)

plt.imshow(resultant_img, 'gray')
plt.axis("off")
plt.show()
```
