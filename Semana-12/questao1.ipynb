```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("imagens/LUA.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, 'gray')
plt.axis("off")
plt.show()
```

```python
#k = [[0, -1, 0] ,[-1, 4, -1], [0, -1, 0]]
k = np.array([[0, -1, 0] ,[-1, 4, -1], [0, -1, 0]]).astype(np.uint8)
laplaciano9 = cv2.Laplacian(img, cv2.CV_64F, k)


plt.imshow(laplaciano9,cmap = 'gray')

```

```python
def corrigirImagem(img):
    maxValue = 0
    minValue = 255
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            maxValue = max(maxValue,img[i][j])
            minValue = min(minValue,img[i][j])
            
    print(maxValue)
    print(minValue)
```

```python
#k = np.array([[-1, -1, -1] ,[-1, 8, -1], [-1, -1, -1]]).astype(np.uint8)
k = np.array([[0, -1, 0] ,[-1, 4, -1], [0, -1, 0]]).astype(np.uint8)
laplaciano = cv2.Laplacian(img, cv2.CV_64F, k).astype(np.uint8)

novaImagem = laplaciano.astype(np.uint8)
novaImagem = cv2.add(img,laplaciano)
#for i in range(laplaciano.shape[0]):
#    for j in range(laplaciano.shape[1]):
#       novaImagem[i][j] = img[i][j] + laplaciano[i][j]
#       #print("{0} = {1} + {2}".format(novaImagem[i][j],int(img[i][j]),int(laplaciano[i][j])))

corrigirImagem(novaImagem)
novaImagem = cv2.blur(novaImagem, (2,2))
plt.imshow(novaImagem,cmap = 'gray')
```
