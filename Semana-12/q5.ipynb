```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("imagens/Fig1026(a)(headCT-Vandy).png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.axis("off")
plt.show()
```

```python
img.shape
```

Borda degrau

```python
img0 = img
buffer = []
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if i == 200: #posicao y
            if j > 380 and j < 410: #posicao x
                buffer.append(img0[i][j])
                img0[i][j] = 0

plt.imshow(img0, cmap='gray')
plt.axis("off")
plt.show()
```

```python
plt.plot(buffer)
ax=plt.gca()
ax.get_xaxis().set_visible(False)
plt.show()
```
