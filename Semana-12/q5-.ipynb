```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagem = cv2.imread("imagens/Fig1026(a)(headCT-Vandy).png", cv2.IMREAD_GRAYSCALE)
plt.imshow(imagem, cmap='gray')
plt.axis("off")
plt.show()
```

```python
imagem.shape
```

```python
def drawLine(imagem,x1,x2,y,corlinha):
    grafico = []
    for i in range(imagem.shape[0]):
        for j in range(imagem.shape[1]):
            if i == y: #posicao y
                if j > x1 and j < x2: #posicao x
                    grafico.append(imagem[i][j])
                    imagem[i][j] = corlinha
                    
    return imagem, grafico
```

TELHADO

```python
imagem, grafico = drawLine(imagem,460,490,310,255)

plt.imshow(imagem, cmap='gray')
plt.axis("off")
plt.show()
```

```python
plt.plot(grafico)
plt.show()
```

```python
imagem, grafico = drawLine(imagem,430,460,190,255)

plt.imshow(imagem, cmap='gray')
plt.axis("off")
plt.show()
```

```python
plt.plot(grafico)
plt.show()
```
